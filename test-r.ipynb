```r
# Create sample data
set.seed(42)
data <- data.frame(
  x = sort(runif(100, 0, 100)),
  y = runif(100, 0, 100)
)
```

```r
head(data)
```

```r
# Create a line plot
plot(data$x, data$y, 
     type = "l", 
     col = "blue",
     xlab = "X",
     ylab = "Y",
     main = "Sample Line Plot")
```
