```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data
data = np.random.rand(100, 2) * 100
data = data[data[:, 0].argsort()]  # Sort by x values

data
```

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(data[:, 0], data[:, 1], 'b-')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Data Line Plot')
plt.show()
```
