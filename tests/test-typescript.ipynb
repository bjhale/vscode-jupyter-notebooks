```typescript
import * as Plot from "npm:@observablehq/plot";
import { DOMParser } from "npm:linkedom";

const document = new DOMParser().parseFromString("<div id='plot'></div>", "text/html");
```

```typescript
const data = Array.from({ length: 100 }, () => ({
  x: Math.random() * 100,
  y: Math.random() * 100,
})).sort((a, b) => a.x - b.x);

data;
```

```typescript
const plot = Plot.plot({
    document,
    marks: [
        Plot.line(data, { x: "x", y: "y" })
    ],
    grid: true
}).toString();

const display = {
    [Symbol.for("Jupyter.display")]() {
        return {
            "image/svg+xml": plot,
        }
    }
};

display;
```
